```python
import pandas as pd

import numpy as np
```

```python
dados_1semestre = 'Preços semestrais - AUTOMOTIVOS_2025.01.csv'
df_1semestre = pd.read_csv(dados_1semestre, sep=';', low_memory=False)
df_1semestre.head(3)
```

```python
dados_2semestre = 'Preços semestrais - AUTOMOTIVOS_2025.02.csv'
df_2semestre = pd.read_csv(dados_2semestre, sep=';')
df_2semestre.head(3)
```

```python
df_2semestre['CNPJ da Revenda'] = df_2semestre['CNPJ da Revenda'].astype
```

```python
df_anp = pd.concat([df_1semestre, df_2semestre], ignore_index=True)
df_anp['Valor de Venda'] = df_anp['Valor de Venda'].str.replace(',','.').astype(float) # .astype(float) para mudar de str para float (tipo de dado)
df_anp
```

```python
df_gasolina = df_anp.loc[df_anp['Produto'] == 'GASOLINA']
df_etanol = df_anp.loc[df_anp['Produto'] == 'ETANOL']
```

```python
df_media_estado_gaso = df_gasolina.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index()
df_media_estado_gaso = df_media_estado_gaso.sort_values(by='Valor de Venda', ascending= False)

df_media_estado_gaso
```

```python
df_media_estado_etanol = df_etanol.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index()
df_media_estado_etanol = df_media_estado_etanol.sort_values(by='Valor de Venda', ascending= False)

df_media_estado_etanol
```

```python
df_media_regiao_gaso = df_gasolina.groupby('Regiao - Sigla')['Valor de Venda'].mean().reset_index()
df_media_regiao_gaso = df_media_regiao_gaso.sort_values(by='Valor de Venda')
df_media_regiao_gaso
```

```python
df_media_regiao_etanol = df_etanol.groupby('Regiao - Sigla')['Valor de Venda'].mean().reset_index()
df_media_regiao_etanol = df_media_regiao_etanol.sort_values(by='Valor de Venda')
df_media_regiao_etanol
```

```python
df_media_munic = df_gasolina.groupby('Municipio')['Valor de Venda'].mean().reset_index()
df_media_munic
```

```python
df_media_munic = df_etanol.groupby('Municipio')['Valor de Venda'].mean().reset_index()
df_media_munic
```
